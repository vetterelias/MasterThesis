```python
#df_containers_wijken = pd.DataFrame(columns=['wijk','year', 'month', 'net_weight','fractie'])


for i in wijken.Wijken:
    containers = df_containers_clean[df_containers_clean.fractie == 'GFT']
    containers_2 = containers[containers.buurt_code.str.match(i)]
    containers_2 = containers_2.groupby(['year','month'])['net_weight'].sum()
    if len(containers[containers.buurt_code.str.match(i)]) >= 1:
        containers_3 = containers_2.reset_index(level=['month','year'])
        containers_3["wijk"]= i
        containers_3["fractie"] = "GFT"
        df_containers_wijken = df_containers_wijken.append(containers_3)
```
